```python
# Region codes need to be sorted so they appear alphabetically
# But the codes occur arbitrarily, so a list of region names should
# First be made into a list, here:
REGION_COD_EQUIV = []
for i in COLLECTIONS:
    REGION_COD_EQUIV.append(i['data']['name'])
    
# Here we remove any undesirable collections from the list
REGION_COD_EQUIV.remove('* Methodological Documents')
REGION_COD_EQUIV.remove('Coin group lack bug')
REGION_COD_EQUIV.remove('Date bug fixing')

# These names are sorted
REGION_COD_EQUIV.sort()

# The JSON code is then queried in order of that list, creating a new
# list of codes that happens to be in alphabetical order
REGION_CODES = []
for i in REGION_COD_EQUIV:
    for j in COLLECTIONS:
        if j['data']['name'] == i:
            REGION_CODES.append(j['key'])
            break
```

```python
# Here we create a dynamic TOC from the REGION_COD_EQUIV list
for i in REGION_COD_EQUIV:
    FILE.write('<a href="#' + i + '">' + i + '</a><br />\n')
#    print('<a href="#' + i + '">' + i + '</a><br />')
```

```python
# This caps of the TOC before getting into regional lit sections
FILE.write("""</p>\n
<p>&nbsp;</p>\n""")
```

```python
# Now we take the codes and iterate on them. Below is the setup for the for loop
# Which constitute the meat of this program. Each REGION_CODE is iterated on
# and then added to the HTML document.
REGIONAL_ITEMS = []
# below rate limiter will be used with the following loop, specifically in the "ITEM" portion
# max_calls is the number made ("max_calls") per time unit ("period") in seconds.
# a second rate limiter ("cool") is used only with errors to let the API cooldown (if this is even
# how the API works)
rate_limiter = RateLimiter(max_calls=1, period=2)
rate_limiter_cool = RateLimiter(max_calls=1, period=10)

for i in REGION_CODES:
# This creates a long JSON file of every region's items (for next loop)
    REGIONAL_ITEMS = zot.collection_items(i)
# This is needed to create the names at the end for the header
    REGIONAL_INFORMATION = zot.collection(i)
    REGIONAL_INFORMATION = REGIONAL_INFORMATION['data']['name']

# This gets a list of item codes from the JSON file   
    ITEM_CODES = []
    for i in REGIONAL_ITEMS:
        if i['data']['itemType'] == 'note' or i['data']['itemType'] == 'attachment':
            continue
        else:
            ITEM_CODES.append(i['key'])

# This queries the Zotero API, with rate limiter, generating the bibliographic entry
# It gets added to a list called ITEM, created at the start of the loop
    ITEM = []
# This code replaces the code below, having added an error loop in 'while True'
    for j in ITEM_CODES:
        with rate_limiter:
            while True:
                try:
                    zot.add_parameters(itemKey=j, content='bib')
                    z = zot.top()
                    ITEM.extend(z)
                except:
                #except SomeSpecificException:
                    with rate_limiter_cool:
                        continue
                break
    
#    for j in ITEM_CODES:
#        with rate_limiter:
#            zot.add_parameters(itemKey=j, content='bib')
#            z = zot.top()
#            ITEM.extend(z)

# Remove default Zotero style elements and replace them with FLAME elements        
    ITEM = [re.sub('<div class="csl-entry">', '<li>', k) for k in ITEM]
    ITEM = [re.sub('</div>', '</li>\n', k) for k in ITEM]
# Sort the list alphabetically (otherwise it appears in arrbitrary order from Zotero)
    ITEM.sort()
# Write the header in for the region.
    FILE.write('<h4 id="' + REGIONAL_INFORMATION + '"><b><u>' + REGIONAL_INFORMATION + '</u></b> <a href="#home">(back)</a></h4>\n<ul>\n')
# Write in every bibliographic item for the region    
    for l in ITEM:
        FILE.write(l)   
# Close up the region with footer
    FILE.write('</ul>\n<div></div>\n')
# After all loops have run, insert all footer material on HTML file    
FILE.write("""			</div>
					</div><!-- /.main -->
			</div><!-- /.row-->
		</div><!-- /.content -->
	</div><!-- /.wrap -->
	<footer id="containerfooter" class="footerclass" itemscope itemtype="https://schema.org/WPFooter">
  <div class="container">
  	<div class="row">
  								        						            </div>
        <div class="footercredits clearfix">
    		
    		        	<p>FLAME and this site (ISSN 2642-911X) are licensed under a Creative Commons Attribution-ShareAlike 3.0 Unported License. 2021 - WordPress Theme by <a href="https://www.kadencewp.com/" rel="nofollow noopener noreferrer" target="_blank">Kadence WP</a></p>
    	</div>

  </div>

</footer>

		</div><!--Wrapper-->
		<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/plugins/counter-number-showcase/assets/js/bootstrap.js?ver=5.7.2' id='wpsm_count_bootstrap-js-front-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/plugins/counter-number-showcase/assets/js/counter_nscript.js?ver=5.7.2' id='wpsm-new_count_script3-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/plugins/counter-number-showcase/assets/js/waypoints.min.js?ver=5.7.2' id='wpsm-new_count_waypoints-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/plugins/counter-number-showcase/assets/js/jquery.counterup.min.js?ver=5.7.2' id='wpsm-new_count_script2-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/themes/virtue/assets/js/min/bootstrap-min.js?ver=3.4.4' id='bootstrap-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-includes/js/hoverIntent.min.js?ver=1.8.1' id='hoverIntent-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-includes/js/imagesloaded.min.js?ver=4.1.4' id='imagesloaded-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-includes/js/masonry.min.js?ver=4.2.2' id='masonry-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/themes/virtue/assets/js/min/plugins-min.js?ver=3.4.4' id='virtue_plugins-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/themes/virtue/assets/js/min/magnific-popup-min.js?ver=3.4.4' id='magnific-popup-js'></script>
<script type='text/javascript' id='virtue-lightbox-init-js-extra'>
/* <![CDATA[ */
var virtue_lightbox = {"loading":"Loading...","of":"%curr% of %total%","error":"The Image could not be loaded."};
/* ]]> */
</script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/themes/virtue/assets/js/min/virtue-lightbox-init-min.js?ver=3.4.4' id='virtue-lightbox-init-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-content/themes/virtue/assets/js/min/main-min.js?ver=3.4.4' id='virtue_main-js'></script>
<script type='text/javascript' src='https://coinage.princeton.edu/wp-includes/js/wp-embed.min.js?ver=5.7.2' id='wp-embed-js'></script>
	</body>
</html>"""
)

FILE.close()
```
